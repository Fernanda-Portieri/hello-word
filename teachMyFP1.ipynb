```python
# Course recommendation

course_data = {
  "Machine Learning Crash Course": "... machine learning algorithms, Python...",
  "Introduction to Data Science": "... data analysis, statistics, Python...",
  # Add more courses and descriptions
}
```

```python
# Recommend courses

def recommend_courses(keywords):
  recommendations = []
  for course_title, description in course_data.items():
    score = sum(word in description for word in keywords)  # Simple matching score
    if score > 0:
      recommendations.append((course_title, score))
  # Sort recommendations by score (descending)
  recommendations.sort(key=lambda x: x[1], reverse=True)
  return recommendations
```
